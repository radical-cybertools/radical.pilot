```python

# lets see how many units are *concurrently* executing, ie. in Executing state.  
# That state is left into the AgentStagingOutputPending state
exe_filter = {'in'  : [{'state' : 'Executing'}],
              'out' : [{'state' : 'AgentStagingOutputPending'}]}

# add a concurrency column with the numbers we look for, and show some of them
rpu.add_concurrency(adv, tgt='cc_exe', spec=exe_filter)
print adv.columns.values
adv[np.isfinite(adv['cc_exe'])][50:70]
```

```python
# on towards plotting that number over time.  
# This is just a helper to get a nice plotting canvas
font = {'family' : 'monospace',
        'weight' : 'normal',
        'size'   : 14}
plt.rc('font', **font)

def create_figure(title=None):
    """
    create a plot object and tune its layout to our liking.
    """
    
    fig, ax = plt.subplots(figsize=(10,5))
    if title:
        ax.set_title(title, fontsize=18)
    
    ax.grid(False)

    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')

    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    ax.spines['bottom'].set_position(('axes', -0.05))
    ax.spines['left'].set_position(('axes', -0.02))
    ax.xaxis.grid(False)
    ax.yaxis.grid(False)

    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    
    fig.tight_layout()

    return ax
```

```python

# plot that concurrency over time

ax_ops = create_figure('Number of concurrently executing CUs over time')
labels = list()

sub_adv = adv[np.isfinite(adv['cc_exe'])]
sub_adv.plot(x='time', y='cc_exe', ax=ax_ops)

labels.append(sid)

ax_ops.set_xlabel("time (s)",               fontsize=14)
ax_ops.set_ylabel("number of executin CUs", fontsize=14)

legend(labels, loc='upper left',          fontsize=12)
savefig('exe_concurrency.png', bbox_inches='tight')
savefig('exe_concurrency.pdf', bbox_inches='tight')
```
